```python
%matplotlib inline
import yfinance as yf
import datetime
import matplotlib.pyplot as plt


start_sp = datetime.datetime(1980, 1, 1)
end_sp = datetime.datetime(2009, 12, 31)


sp500 = yf.download('^GSPC',start_sp,end_sp)
```

```python
sp500['Open'].plot()
plt.tight_layout()
plt.title('Open Price')
plt.grid()
plt.show()
```

```python
ndx = yf.download('^NDX', start_sp,end_sp)
    

ndx['Open'].plot()
plt.tight_layout()
plt.title('Open Price')
plt.grid()
plt.show()
```

```python
hsi = yf.download('^HSI', start_sp,end_sp)
    

hsi['Open'].plot()

plt.tight_layout()
plt.title('Open Price')
plt.grid()
plt.show()
```

```python
ndx['Open'].plot()
sp500['Open'].plot()

plt.tight_layout()
plt.title('Open Price')
plt.grid()
plt.show()
```

```python
def daily_change(x):
    return x.shift(-1) / x -1

change_sp500 = daily_change(sp500['Close'])
change_ndx = daily_change(ndx['Close'])
change_hsi = daily_change(hsi['Close'])

change_sp500.plot()
change_ndx.plot()
change_hsi.plot()

plt.tight_layout()
plt.title('Change Index')
plt.grid()
plt.show()
```
